# The following data is a copy of the steam database from 1/28/2019
It was scraped with python, fed as a sqlite db into a csv converter, and processed for analysis

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('games.csv', error_bad_lines=False, sep=';')
```

```python
df.shape
```

```python
df.head()
```

```python
df['Price'] = df['Price'] / 100
```

```python
df.head()
```

```python
df2 = pd.read_csv('genre_count.csv', error_bad_lines=False, sep=';')
df3 = pd.read_csv('unique_devs.csv', error_bad_lines=False, sep=';')
df4 = pd.read_csv('hidden_games.csv', error_bad_lines=False, sep=';')
df5 = pd.read_csv('publishers.csv', error_bad_lines=False, sep=';')
df6 = pd.read_csv('metacritic_scores.csv', error_bad_lines=False, sep=';')
```

```python
df_final = pd.concat([df, df2, df3, df4, df5, df6], ignore_index=True)
```
